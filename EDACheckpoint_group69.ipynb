# Research Question

What is the correlation between the average temperature of a given state and the quantity of Division I NCAA basketball players generated between season 2018-2019 and 2023-2024? Furthermore, how does the average temperature per state relate to the skill level of Division I basketball players, as indicated by the average of points, rebounds, and assists over the season, within that same timeframe?

## Background and Prior Work

The debate over which state produces the best basketball players has long been a point of contention among Americans, with popular contenders including New York, California, and Florida. While previous research has attempted to answer this question, it has only focused on the socio-economic factors, such as the study by James Tompsett and Chris Knoester that focused extensively on the likelihood of high school athletes advancing to collegiate levels.<a name="cite_ref_1"/>[<sup>1</sup>](#cite_note_1) Indicators of family socioeconomic statuses, athletic development and merit, academic expectations and knowledge, and school contexts were used to predict the likelihood of becoming a college athlete. It was shown that high school athletes with a stronger socioeconomic background are much more likely to become collegiate athletes. The correlation with this is clear but it still does not answer our question. Thus, we aim to investigate another less conventional variable that could help us identify the best basketball state - the average temperature of the state.

There have been studies in the past that have looked at the effect of environmental factors on active participation in physical activities. The paper titled "Assessing the Effects of Weather Conditions on Physical Activity Participation Using Objective Measures" by Catherine B. Chan and Daniel A. Ryan examines how different weather conditions influence physical activity levels.<a name="cite_ref_2"/>[<sup>2</sup>](#cite_note_2) The study identifies weather as a significant barrier to physical activity, particularly focusing on how adverse conditions such as low temperatures, rain, and snow decrease participation. Using data from various demographic groups, the study highlights that physical activity levels are generally higher on warmer days and lower during inclement weather. We intend on extending this study into the context of basketball. Further, a study done by David Hancock, Matthew Vierimaa, and Ashley Newman revealed that communities with populations of 250,000-499,999 are talent hotspots for high-level competitive basketball. They explained that this is due to favorable infrastructure and social structure, but call for more research into other factors such as crime rates and green spaces. <a name="cite_ref_3"/>[<sup>3</sup>](#cite_note_3)

We aim to examine the correlation between a state's average temperature and the number of Division I NCAA basketball players it generates between 2019 and 2024 and foresee that environmental factors such as average temperature would provide residents in those areas different, unequal opportunities for physical activity. Another factor that could come into play is sports culture. As sports are strongly integrated in cultures, we may also observe a change in the popularity of different sports due to weather and how accessible that sport is. The accessibility of a sport due to climate could potentially have an effect on the opportunities for training and skill development, further impacting on court performance down the line. Investigating how average temperatures in different states correlate with the emergence of Division I basketball players can offer valuable insights into the broader determinants of athletic success.

```python
import pandas as pd
```

Now let's merge these values for temperature and count for individual states and analyze if there is a mathematical correlation to support this visual

```python
# Renaming columns to match for merging
player_bar_df.rename(columns={'State': 'state'}, inplace=True)

# Merging the dataframes on the 'state' column
merged_df = pd.merge(bar_df, player_bar_df, on='state', how='outer')

merged_df.head()
```

```python
sns.scatterplot(x='average_temperature', y='Count', data=merged_df, s=100)
plt.xlabel('Average Temperature')
plt.ylabel('Player Count');
```

This scatter plot seems to have a positive relationship, but let's draw a line through to make sure.

```python
# Converting all values in 'average_temperature' and 'Count' columns to float
merged_df['average_temperature'] = merged_df['average_temperature'].astype(float)
merged_df['Count'] = merged_df['Count'].astype(float)

sns.lmplot(x='average_temperature', y='Count', data=merged_df);
```

This seems to be a relatively positive slope, but we can use linear regression to check our hypothesis.

```python
# Statmodels & patsy
import patsy
import statsmodels.api as sm
import statsmodels.formula.api as smf

import warnings
warnings.filterwarnings('ignore')

from scipy.stats import skewnorm

#improve resolution
#comment this line if erroring on your machine/screen
%config InlineBackend.figure_format ='retina'

outcome, predictors = patsy.dmatrices('Count ~ average_temperature', merged_df)
model = sm.OLS(outcome, predictors)
```

```python
## fit the model
results = model.fit()

## look at the results
print(results.summary())
```

The model shows that there is a statistically significant positive relationship between average_temperature and Count. Specifically, as the average temperature of a state increases, the player count in a state also tends to increase. This coefficient for average_temperature means that for each additional degree increase in average_temperature, the Count is expected to increase by approximately 5.82 players. However, the R-squared value is relatively low (0.264), indicating that while the relationship is significant, average temperature only explains about 26.4% of the variance in player count. This suggests that other factors also play a significant role in determining the player count. These factors could include the accessibility to facilities with basketball courts or family income in certain scenarios.

Now let's merge these values for temperature and average ppg for individual states and analyze if there is a mathematical correlation to support this visual

```python
# Merging the weather and ppg dataframes on the 'state' column
ppg_weather = pd.merge(bar_df, overall_avg_ppg, on='state', how='outer')

# Drop states without PPG data
ppg_weather.dropna(how="any", inplace=True)

ppg_weather.head()
```

```python
# Plot weather vs ppg
sns.scatterplot(x='average_temperature', y='PPG', data=ppg_weather, s=100)
plt.title('Average Temperature vs Average Points per Game')
plt.xlabel('Average Temperature')
plt.ylabel('Points per Game');
```

It's hard to tell the relationship from the plot so let's add a line for it

```python
# Add linear regression line
# Converting all values in 'average_temperature' and 'Count' columns to float
ppg_weather['average_temperature'] = ppg_weather['average_temperature'].astype(float)
ppg_weather['PPG'] = ppg_weather['PPG'].astype(float)

sns.lmplot(x='average_temperature', y='PPG', data=ppg_weather);
plt.title('Average Temperature vs Average Points per Game')
plt.xlabel('Average Temperature (Farenheit)')
plt.ylabel('Average Points per Game')
```

It looks like we we have negative relationship between average temperature and average point per game. In order to strongly support our hypothesis, let's use the OLS model

```python
# OLS analysis
outcome, predictors = patsy.dmatrices('PPG ~ average_temperature', ppg_weather)
model = sm.OLS(outcome, predictors)

result = model.fit()
print(result.summary())
```

The R-squared value is 0.055, indicating that the model explains only around 5.5% of the variation in PPG using average temperature. This suggests a poor fit and a weak relationship between the two variables..

The coefficient for average_temperature (-0.0382) indicates a negative relationship, meaning that for every one-unit increase in average temperature, PPG is predicted to decrease by 0.0382 units, on average. However, the p-value of 0.120 for this coefficient suggests that it is not statistically significant.

Overall, the regression results suggest that average temperature is not a strong predictor of Points Per Game, as the model has a low R-squared value and the temperature coefficient is not statistically significant.
